```python
np.random.seed(1)

# Class means
a_means = [np.array([-8.0, -6.0]), np.array([2.0, -12.0])] 
b_means = [np.array([10.0, 5.0]), np.array([0.0, 12.0])]

# Class covariances
covs = [np.array([[5, 0], [0, 5]]), np.array([[4, 0], [0, 4]])]

# Size of the larger peak
p = 0.85

_ = generate_dataset(a_means, covs, b_means, covs, num_samples=400,
                     modality=[p, 1-p], file_name="../datasets/bimodal_dataset_1.csv")
```
